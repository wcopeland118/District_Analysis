# PyCity Schools Analysis

## Setup

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
df.head()
```

```python
df.shape
```

```python
df.count()
```

```python
df = df.rename(columns={"Student ID":"student_id", "School ID":"school_id"})
df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
df.head()
```

```python
#total number of schools
total_schools = df["school_id"].nunique()
total_schools
```

```python
#total number of students
total_students = df["student_id"].nunique()
total_students
```

```python
#total budget
total_budget = df["budget"].unique().sum()
total_budget
```

```python
#average math score
average_math = df["math_score"].mean()
average_math
```

```python
#average reading score
average_reading = df["reading_score"].mean()
average_reading
```

```python
#overall passing rate
passing_math = len(df.loc[df["math_score"]>=70,:])
passing_reading = len(df.loc[df["reading_score"]>=70,:])
overall_passing = (passing_math+passing_reading)/(total_students*2)
overall_passing
```

```python
#overall average
df["overall_average"] = (df["math_score"]+df["reading_score"])/2
df["overall_average"].mean()
```

```python
#overall average passing rate
overall_average_passing = len(df.loc[df["overall_average"]>=70,:])/total_students
overall_average_passing
```

```python
#percentage of students with a passing math score (70 or greater)
percent_math = df.loc[df["math_score"]>=70,:]["student_id"].count()/total_students
percent_math
```

```python
#percentage of students with a passing reading score (70 or greater)
percent_reading = df.loc[df["reading_score"]>=70,:]["student_id"].count()/total_students
percent_reading
```

```python
#dataframe to hold the above results
df_totals = pd.DataFrame({"Total Schools": total_schools,
                   "Total Students": f"{total_students:,}",
                   "Total Budget": f"${total_budget:,.2f}",
                   "Average Math Score": f"{average_math:.2f}",
                   "Average Reading Score": f"{average_reading:.2f}",
                   "Percent Passing Math": f"{percent_math:.2%}",
                   "Percent Passing Reading": f"{percent_reading:.2%}",
                   "Overall Passing Rate": f"{overall_passing:.2%}"
                    },index = [0])
df_totals
```
